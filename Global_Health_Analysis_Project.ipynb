**Outliers Detection**

## Identify Treatment Cost Outliers


````python
Q1 = df['Average Treatment Cost (USD)'].quantile(0.25)
Q3 = df['Average Treatment Cost (USD)'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
````

````python
upper_bound_extreme = Q3 + (3 * IQR)
lower_bound_extreme = Q1 - (3 * IQR)

print(f"Upper Bound for Extreme Outliers: {upper_bound_extreme}")
print(f"Lower Bound for Extreme Outliers: {lower_bound_extreme}")

treatment_cost_outliers = df[(df['Average Treatment Cost (USD)'] > upper_bound_extreme) | (df['Average Treatment Cost (USD)'] < lower_bound_extreme)]

print("\nExtreme Outliers in 'Average Treatment Cost (USD)':")
display(treatment_cost_outliers[['Disease Name', 'Average Treatment Cost (USD)']])
````

## Identify DALYs Outliers


````python
Q1_dalys = df['DALYs'].quantile(0.25)
Q3_dalys = df['DALYs'].quantile(0.75)
IQR_dalys = Q3_dalys - Q1_dalys

print(f"Q1 for DALYs: {Q1_dalys}")
print(f"Q3 for DALYs: {Q3_dalys}")
print(f"IQR for DALYs: {IQR_dalys}")
````

````python
upper_bound_extreme_dalys = Q3_dalys + (3 * IQR_dalys)
lower_bound_extreme_dalys = Q1_dalys - (3 * IQR_dalys)

print(f"Upper Bound for Extreme DALYs Outliers: {upper_bound_extreme_dalys}")
print(f"Lower Bound for Extreme DALYs Outliers: {lower_bound_extreme_dalys}")

dalys_outliers = df[(df['DALYs'] > upper_bound_extreme_dalys) | (df['DALYs'] < lower_bound_extreme_dalys)]

print("\nExtreme Outliers in 'DALYs':")
display(dalys_outliers[['Country', 'DALYs']])
````

## Identify Mortality Rate Outliers


````python
Q1_mortality = df['Mortality Rate (%)'].quantile(0.25)
Q3_mortality = df['Mortality Rate (%)'].quantile(0.75)
IQR_mortality = Q3_mortality - Q1_mortality

print(f"Q1 for Mortality Rate (%): {Q1_mortality}")
print(f"Q3 for Mortality Rate (%): {Q3_mortality}")
print(f"IQR for Mortality Rate (%): {IQR_mortality}")

upper_bound_extreme_mortality = Q3_mortality + (3 * IQR_mortality)
lower_bound_extreme_mortality = Q1_mortality - (3 * IQR_mortality)

print(f"Upper Bound for Extreme Mortality Rate (%) Outliers: {upper_bound_extreme_mortality}")
print(f"Lower Bound for Extreme Mortality Rate (%) Outliers: {lower_bound_extreme_mortality}")

mortality_rate_outliers = df[(df['Mortality Rate (%)'] > upper_bound_extreme_mortality) | (df['Mortality Rate (%)'] < lower_bound_extreme_mortality)]

print("\nExtreme Outliers in 'Mortality Rate (%):'")
display(mortality_rate_outliers[['Country', 'Disease Category', 'Mortality Rate (%)']])
````

## Summary of Outlier Analysis

Based on the Interquartile Range (IQR) method (3 * IQR for extreme outliers):

- **Average Treatment Cost (USD)**: No extreme outliers were identified. This suggests that while there might be some variability, there are no data points that fall drastically outside the typical range as defined by the 3 * IQR rule.

- **DALYs (Disability-Adjusted Life Years)**: No extreme outliers were identified. This indicates that the DALYs values are relatively consistent across the dataset and do not contain any data points that are exceptionally high or low compared to the majority.

- **Mortality Rate (%)**: No extreme outliers were identified. Similar to the other metrics, the mortality rates appear to be within an expected distribution without extreme deviations.

In conclusion, the initial extreme outlier detection using the 3 * IQR rule did not find any data points that were considered extreme outliers for 'Average Treatment Cost (USD)', 'DALYs', or 'Mortality Rate (%)' in this dataset. The box plots visually confirm this by showing no individual data points beyond the computed extreme outlier bounds.

**Disease-Specific Trends**

## Identify Diseases with Greatest Mortality Reduction


````python
disease_mortality_trends = df.groupby(['Disease Name', 'Year'])['Mortality Rate (%)'].mean().reset_index()

disease_mortality_reduction = []

for disease_name in disease_mortality_trends['Disease Name'].unique():
    disease_data = disease_mortality_trends[disease_mortality_trends['Disease Name'] == disease_name]

    if len(disease_data) > 1: # Only consider diseases with data for more than one year
        earliest_year_data = disease_data.sort_values(by='Year').iloc[0]
        latest_year_data = disease_data.sort_values(by='Year').iloc[-1]

        earliest_mortality = earliest_year_data['Mortality Rate (%)']
        latest_mortality = latest_year_data['Mortality Rate (%)']

        mortality_reduction = earliest_mortality - latest_mortality

        disease_mortality_reduction.append({
            'Disease Name': disease_name,
            'Earliest Year Mortality (%)': earliest_mortality,
            'Latest Year Mortality (%)': latest_mortality,
            'Mortality Reduction (%)': mortality_reduction
        })

df_mortality_reduction = pd.DataFrame(disease_mortality_reduction)

print("Mortality Reduction Calculation for Each Disease (first 10 rows):")
display(df_mortality_reduction.head(10))
````

````python
top_10_reductions = df_mortality_reduction.sort_values(by='Mortality Reduction (%)', ascending=False).head(10)

print("Top 10 Diseases with Greatest Mortality Reduction (%):")
display(top_10_reductions)
````
